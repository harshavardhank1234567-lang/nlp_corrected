```python
# Program 1



import nltk

# Download required datasets
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab') # Added to fix the LookupError

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer

# Sample text
text = "Natural language processing includes tokenization, removing stopwords, stemming, and lemmatization."

# 1. Tokenization
tokens = word_tokenize(text)
print("Tokens:", tokens)

# 2. Stop word removal
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
print("After Stop Word Removal:", filtered_tokens)
         # 3. Stemming
ps = PorterStemmer()
stemmed = [ps.stem(word) for word in filtered_tokens]
print("Stemmed:", stemmed)

# 4. Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized = [lemmatizer.lemmatize(word) for word in filtered_tokens]
print("Lemmatized:", lemmatized)
```

```python
#Program 2

import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer

# Download required NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')

# Sample text
text = """Natural language processing includes tokenization, removing stopwords, stemming, and lemmatization.
It helps machines understand human language effectively."""

# Tokenization
tokens = word_tokenize(text)

# Remove stopwords and punctuation
stop_words = set(stopwords.words('english'))
clean_tokens = [w for w in tokens if w.lower() not in stop_words and w.isalpha()]

# Stemming and Lemmatization
ps = PorterStemmer()
lemmatizer = WordNetLemmatizer()

print("\nWord Transformations:\n")
print(f"{'Original':<20} {'Stemmed':<20} {'Lemmatized'}")
print("-" * 55)

for word in clean_tokens:
    print(f"{word:<20} {ps.stem(word):<20} {lemmatizer.lemmatize(word)}")
```

```python
# Program 4

import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer

# Download required resources (run once)
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')

# Sample movie reviews (small corpus)
reviews = [
    "The movie was fantastic! I loved the thrilling scenes and amazing acting.",
    "It was boring and too long. I didn't like the predictable plot.",
    "Absolutely wonderful cinematography and great story.",
]

# Initialize stemmer and lemmatizer
ps = PorterStemmer()
lemmatizer = WordNetLemmatizer()

for i, review in enumerate(reviews, 1):
    print(f"\n--- Review {i} ---")
    print("Original:", review)

    tokens = word_tokenize(review)

    stemmed = ' '.join(ps.stem(word) for word in tokens)
    lemmatized = ' '.join(lemmatizer.lemmatize(word) for word in tokens)

    print("Stemmed:   ", stemmed)
    print("Lemmatized:", lemmatized)
```

```python
# program 5

import nltk
from nltk import word_tokenize, pos_tag, ne_chunk
from nltk.tree import Tree
from sklearn.metrics import precision_recall_fscore_support

nltk.download('maxent_ne_chunker')
nltk.download('maxent_ne_chunker_tab')
nltk.download('words')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger_eng')

data = [
    ("Barack Obama was born in Hawaii.", ["Barack Obama"]),
    ("Apple is looking at buying U.K. startup for $1 billion.", []),
    ("Elon Musk is the CEO of SpaceX.", ["Elon Musk"]),
    ("Angela Merkel met with Vladimir Putin in Moscow.", ["Angela Merkel", "Vladimir Putin"]),
]

def extract_entities(text):
    return [" ".join(w for w, _ in c.leaves())
            for c in ne_chunk(pos_tag(word_tokenize(text)))
            if isinstance(c, Tree) and c.label() == "PERSON"]

y_true = [set(t) for _, t in data]
y_pred = [set(extract_entities(s)) for s, _ in data]

true_all = {e for s in y_true for e in s}
pred_all = {e for s in y_pred for e in s}
all_entities = list(true_all | pred_all)

precision, recall, f1, _ = precision_recall_fscore_support(
    [1]*len(all_entities),
    [1 if e in pred_all else 0 for e in all_entities],
    average='micro'
)

for i, (s, t) in enumerate(data):
    print(f"\nSentence: {s}")
    print(f"True: {t}")
    print(f"Pred: {list(y_pred[i])}")

print(f"\nPrecision: {precision:.2f}, Recall: {recall:.2f}, F1: {f1:.2f}")
```

```python
# Program 6

import re
from collections import defaultdict

text = """Once upon a time there was a clever fox in a forest.
The fox was hungry and looking for food.
He saw a crow with cheese in her beak.
The fox made a plan to get the cheese.
He praised the crow and asked her to sing.
The cheese fell and the fox ate it."""

words = re.sub(r'[^\w\s]', '', text.lower()).split()

def build(n):
    m = defaultdict(lambda: defaultdict(int))
    for i in range(len(words)-n):
        m[tuple(words[i:i+n])][words[i+n]] += 1
    return m

predict = lambda m, c: max(m.get(tuple(c), {}), key=m.get(tuple(c), {}).get, default=None)

def acc(m, n):
    return sum(predict(m, words[i:i+n]) == words[i+n] for i in range(len(words)-n)) / (len(words)-n)

while True:
    n = input("\nN (1/2/3) or exit: ")
    if n == 'exit': break
    if not n.isdigit() or int(n) < 1:
        print("Invalid"); continue

    n, m = int(n), build(int(n))
    print("Accuracy:", round(acc(m, n), 2))

    while True:
        c = input(f"{n} word(s) or back: ").lower()
        if c in ['back', 'exit']:
            if c == 'exit': exit()
            break
        c = c.split()
        if len(c) != n:
            print(f"Enter {n} words"); continue

        print("Next:", predict(m, c) or "No prediction")
```

```python
# Program 7

import random
from collections import defaultdict

text = """Because I could not stop for Death –
He kindly stopped for me –
The Carriage held but just Ourselves And Immortality.
We slowly drove He knew no haste
And I had put away
My labor and leisure too
For His Civility –
We passed the School where Children strove
At Recess in the Ring
We passed the Fields of Gazing Grain
We passed the Setting Sun"""

def build(text, o=2):
    w = text.replace("–", "").split()
    m = defaultdict(list)
    for i in range(len(w)-o): m[tuple(w[i:i+o])].append(w[i+o])
    return m, set(w)

def gen(m, o=2, n=16, l=4):
    p = list(random.choice(list(m.keys())))
    for _ in range(n-o):
        k = tuple(p[-o:])
        if k not in m: break
        p.append(random.choice(m[k]))
    return "\n".join(" ".join(p[i:i+l]) for i in range(0,len(p),l)), p

def score(p, c):
    s = set(p)
    pv = {"Death","Immortality","Soul","Heaven","Carriage","Civility","Sun","Grain","Ring"}
    return round((0.7*len(s&c)/len(s) + 0.3*len(s&pv)/len(s))*100, 2) if s else 0

chain, corpus = build(text)

print("Generated Poems:\n")
for i in range(10):
    t, w = gen(chain)
    print(f"Poem {i+1} ({score(w, corpus)}%):\n{t}\n")
```

```python
# Program 8

import pandas as pd, re
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Load & clean
df = pd.read_csv("movie_review.csv")
df.columns = df.columns.str.strip().str.lower()
df.dropna(subset=['review','rating'], inplace=True)

clean = lambda t: re.sub(r"\s+", " ", re.sub(r"\d+|[^\w\s]", "", t.lower())).strip()
X, y = df['review'].apply(clean), df['rating']

# Split + TF-IDF
Xtr, Xte, ytr, yte = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
tfidf = TfidfVectorizer(max_features=5000, ngram_range=(1,2), stop_words='english')
Xtr, Xte = tfidf.fit_transform(Xtr), tfidf.transform(Xte)

# Train & evaluate
for name, model in {"Naive Bayes": MultinomialNB(), "SVM": LinearSVC()}.items():
    model.fit(Xtr, ytr)
    pred = model.predict(Xte)
    print(f"\n=== {name} ===")
    print("Accuracy:", accuracy_score(yte, pred))
    print("Report:\n", classification_report(yte, pred, zero_division=0))
    print("Confusion:\n", confusion_matrix(yte, pred))
```

```python
# Program 9

import pandas as pd

df = pd.read_csv('Restaurant reviews.csv')
df = df.dropna(subset=['Review'])
df = df[df['Review'].str.len() > 0]
print(f"Loaded {len(df)} reviews")

aspects = {
    'food': ['food','biryani','pasta','delicious','soup','corn','fish','paneer','sizzler'],
    'service': ['service','staff','waiter','courteous','pradeep','papiya','polite','prompt'],
    'ambiance': ['ambience','ambiance','atmosphere','place','music','lighting']
}

pos_words = ['good','great','excellent','amazing','delicious','nice','awesome','fantastic']
neg_words = ['bad','poor','disappointed','terrible','stale','impolite']

def analyze(t):
    t = t.lower()
    res = {}
    for a, k in aspects.items():
        if any(x in t for x in k):
            s = [i for i in t.split('.') if any(x in i for x in k)]
            p = sum(w in i for i in s for w in pos_words)
            n = sum(w in i for i in s for w in neg_words)
            res[a] = 'positive' if p > n else 'negative' if n > p else 'neutral'
    return res

print("\n=== SAMPLE ===")
for _, r in df.head(3).iterrows():
    print(f"\n{r['Review'][:80]}...\nRating: {r['Rating']}")
    for a, s in analyze(r['Review']).items():
        print(f"{a}: {s}")

summary = {}
for r in df['Review']:
    for a, s in analyze(r).items():
        summary.setdefault(a, {'positive':0,'negative':0,'neutral':0})[s] += 1

print(f"\n=== SUMMARY ({len(df)} reviews) ===")
for a, v in summary.items():
    tot = sum(v.values())
    print(f"{a}: {tot}")
    for k, c in v.items():
        print(f"  {k}: {c} ({c/tot*100:.0f}%)")
```

```python
# Program 10

import numpy as np
from scipy.spatial.distance import cosine

np.random.seed(42)
words = ['king','queen','man','woman','paris','france','london','england',
         'cat','dog','happy','sad','big','small','hot','cold']

# Create embeddings
w2v = {w: np.random.randn(100) for w in words}
glove = {w: np.random.randn(100) for w in words}

# Enhance relations
def enhance(e):
    e['queen']=e['king']+(e['woman']-e['man'])+np.random.randn(100)*0.1
    e['france']=e['paris']+np.random.randn(100)*0.2
    e['england']=e['london']+np.random.randn(100)*0.2
    e['dog']=e['cat']+np.random.randn(100)*0.3
    e['sad']=-e['happy']+np.random.randn(100)*0.2
    e['small']=-e['big']+np.random.randn(100)*0.2
    e['cold']=-e['hot']+np.random.randn(100)*0.2

enhance(w2v); enhance(glove)

sim = lambda e,a,b: 1 - cosine(e[a], e[b])
ana = lambda e,a,b,c: max(((w,1-cosine(e[b]-e[a]+e[c],e[w])) for w in e if w not in [a,b,c]), key=lambda x:x[1])

pairs = [('king','queen'),('man','woman'),('paris','france'),('london','england'),
         ('cat','dog'),('happy','sad'),('big','small'),('hot','cold'),
         ('king','man'),('queen','woman')]

analogies = [('king','queen','man'),('paris','france','london'),
             ('big','small','hot'),('happy','sad','big')]

print("SIMILARITY:")
w2, g = [], []
for a,b in pairs:
    s1,s2 = sim(w2v,a,b), sim(glove,a,b)
    w2.append(s1); g.append(s2)
    print(f"{a}-{b}: {s1:.2f} | {s2:.2f}")

print(f"\nAvg: W2V={np.mean(w2):.2f}, GloVe={np.mean(g):.2f}")

print("\nANALOGIES:")
for a,b,c in analogies:
    print(f"{a}:{b}::{c}:? → {ana(w2v,a,b,c)[0]} | {ana(glove,a,b,c)[0]}")

print("\nWord2Vec: local context | GloVe: global stats")
```
